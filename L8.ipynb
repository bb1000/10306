### matrix muliplication

Straight forward looping is slow in Python and scales as $N^3$

```python
import time

n = 512
a = numpy.ones((n, n))
b = numpy.ones((n, n))
c = numpy.zeros((n, n))
t1 = time.time()
for i in range(n):
    for j in range(n):
        for k in range(n):
            c[i, j] += a[i, k]*b[k, j]
t2 = time.time()
print("Loop timing", t2-t1)
print(c)
```

The numpy implementation uses highly optimized code

```python
t1 = time.time()
c = a @ b
t2 = time.time()
print("dot timing", t2-t1)
print(c)
```
