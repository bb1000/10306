## Visualization

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(-np.pi, np.pi, 100)
#x
```

```python
y1 = np.sin(x)
y2 = np.cos(x)
```

```python
plt.plot(x, y1)
plt.plot(x, y2)
plt.show()
```

```python
plt.plot(x, y1, 'bo')
plt.plot(x, y2, 'r+')
plt.show()
```

```python
plt.show() # previous plot has been cleard
```

```python
# modify axes
plt.plot(x, y1)
plt.plot(x, y2)
plt.xticks([-np.pi, 0, np.pi], [r'$-\pi$', '0', r'$\pi$']) # raw string
plt.yticks([-1, 0, 1])
plt.show()
```

```python
# add legends
plt.plot(x, y1, label='sin(x)')
plt.plot(x, y2, label='cos(x)')
plt.xticks([-np.pi, 0, np.pi], [r'$-\pi$', '0', r'$\pi$']) # raw string
plt.yticks([-1, 0, 1])
plt.legend()
plt.show()
```

```python
# add some title text
plt.plot(x, y1, label='sin(x)')
plt.plot(x, y2, label='cos(x)')
plt.xticks([-np.pi, 0, np.pi], [r'$-\pi$', '0', r'$\pi$']) # raw string
plt.yticks([-1, 0, 1])
plt.legend()
plt.title('Examples from trigonometry')
plt.xlabel('Angle in radians')
plt.ylabel('Elementry function values')
plt.show()
```
