#### Person update: str vs repr

```python
class Person:
    count = 0  # class attribute
    def __init__(self, given_name, surname):
        self.given_name = given_name  # instance attribute
        self.surname = surname        # instance attribute
        Person.count += 1
        self.number = Person.count
    def __str__(self):
        return f"Person: {self.given_name} {self.surname}"
    def __repr__(self):
        return f'Person("{self.given_name}", "{self.surname}")'
```

```python
p = Person("Jane", "Doe")
```

```python
p
```

```python
repr(p)
```

```python
str(p)
```

```python
p.surname
```

```python
p.__getattribute__('surname')
```

```python
p.__dict__
```

```python
p.__dict__['surname']
```

### Example: complex numbers

```python
class MyComplex: # name in CamelCase style
    def __init__(self, real=0.0, imaginary=0.0):
        self.real = float(real)
        self.imaginary = float(imaginary)

    def __repr__(self):
        return f'MyComplex({self.real}, {self.imaginary})'

    def __str__(self):
        return f'{self.real} + {self.imaginary}i'

    def __add__(self, other): # defines the expression self + other
        z1 = self
        z2 = other
        z = MyComplex(z1.real + z2.real, z1.imaginary + z2.imaginary)
        return z
```

```python
# compare float
float(), float(12), float('14.7')
```

```python
z = MyComplex()
z
```

```python
z.real
```

```python
z.imaginary
```

```python
z.__dict__
```

```python
print(z)
```

```python
z1 = MyComplex(1, 2)
z2 = MyComplex(3, 4)
z1
```

```python
z = z1 + z2
z
```

```python
print(z)
```
