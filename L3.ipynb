Observation:

* viewing `isnumeric` as a normal function a string as a first argument

```python
str.isnumeric('314')
```

* common practice is to use class-based (more on that later) calls, where the dot follows the value and the function comes next

```python
'314'.isnumeric()
```

* in both cases the value `314` is passed as input to the function `isnumeric` as first argument
* in the latter case it is convenient to pass data from one function to next in a chain

~~~
' Hello world\n' -> strip -> 'Hello world' -> lower -> 'hello world'
~~~

```python
' Hello world\n'.strip()
```

```python
' Hello world\n'.strip().lower()
```

```python
help(str.lower)
```

#### combine numbers and text

e.g. printing "7 * 8 = \<value\> " where we insert the value of the operation

```python
7 * 8
```

```python
print('7 * 8 =', 7 * 8)
```

```python
result = '7 * 8 =' + 7 * 8
```

```python
result = '7 * 8 =' + str(7 * 8)
print(result)
```

```python
# f-strings: putting values into string
result = f'7 * 8 = {7 * 8}'
print(result)
```

```python
# str.format - if don't have the final value yet
result = '7 * 8 = {}'
# {} marks a placefolder for some value
result.format(7 * 8)
```
