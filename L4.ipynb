## Loops

```python
a_list
```

```python
# do somethine with each element
print(a_list[0])
print(a_list[1])
print(a_list[2])
```

```python
# the changeing part put in a variable
for i in [0, 1, 2]:
    print('i=', i)
    print(a_list[i])
```

```python
for member in a_list:
    print('member=', member)
```

```python
for member in a_tuple:
    print('member=', member)
```

```python
for member in newdict:
    print('member=', member)
```

```python
for member in newdict.items():
    print('member=', member)
```

```python
# common for dicts
```

```python
for key, value in newdict.items():
    print('key=', key, 'value=', value)
```
