### Scope

* variables at modlue level have global scope
* functions can access those values

```python
a = 13 #global variable

def f():
    a = 7  #local variable
    print(a)

f()
print(a)
```

```python
a = 13 #global variable

def f():
    global a  #usually not recommended practise
    a = 7  #local variable
    print(a)

f()
print(a)
```

### Type hints

* dynmaically typed language
* variable can change type during program
* type hints used to improve documentation

```python
# Example: countdown
import time

def countdown(start: int) -> None:
    """
    Count down in second from start to 0

    paramters: start (int) > 0
    """
    
    def report():
        print(n)

    def decrement():
        nonlocal n  # look for definition outside function decrement
        n = n - 1

    n = start

    while n > 0: # simplified loop
        #print(n)
        report()
        decrement()
        time.sleep(1)

countdown(3)      
```

* `*args` means any number of arguments / collected in a tuple

```python
import sys
def myprint(*args, sep=' ', end='\n'):
    print(args)
    to_terminal = sep.join(str(a) for a in args)
    sys.stdout.write(to_terminal + end)
```

```python
myprint("Hello", 1, [])
```

```python
myprint("Hello", 1, [], sep='\n\n')
```

```python
myprint("Hello", 1, [], sep='\n\n', end='Done')
```

```python
def myprint2(*args, sep=' ', end='\n'):
    print(*args) # equivalent to print(args[0], args[1], ...)

myprint2("Hello", "world")
```

```python
# arbitrary keyword arguments

def myprint3(*args, **kwargs):
    print(args, kwargs)
    print(*args, **kwargs)
```

```python
myprint3()
```

```python
myprint3('hello', 1, sep='|') # inside it will call print('hello', 1, sep='|')
```

* `kwargs` collects input keyed argumetns into a dictionary
* can be passed on to other functions with same notation `**kwargs`
