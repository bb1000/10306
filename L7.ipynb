### functools

https://docs.python.org/3/library/functools.html

```python
import time
import math
def slow_function(x):
    time.sleep(3)
    return math.sqrt(x)
```

```python
%%time
slow_function(2)
```

```python
import functools
cached_function = functools.cache(slow_function)
```

```python
%%time
cached_function(2)
```

```python
%%time
cached_function(2)
```

### argparse

https://docs.python.org/3/library/argparse.html

* approach 1

```python
%%file saving.py
"""
Calcluates savings over several years given interest rate and number of years
"""

import sys

if len(sys.argv) > 2:
    interest_rate = float(sys.argv[1])
    number_of_years = int(sys.argv[2])
else:
    print(f"Usage: {sys.argv[0]} interest_rate number_of_years")
    exit()

amount = 1000

for i, y in enumerate(range(number_of_years), start=1):
    amount += interest_rate/100*amount
    amount = round(amount, 2)
    print(f'Savings after {i:2d} years: {amount:.2f}')
```

```python
!python saving.py 10 10
```

* approach 2

```python
%%file saving2.py
"""
Calcluates savings over several years given interest rate and number of years
"""

import argparse
parser = argparse.ArgumentParser()

parser.add_argument('interest_rate', type=float)
parser.add_argument('number_of_years', type=int)

args = parser.parse_args()                    
print(args)
number_of_years = args.number_of_years
interest_rate = args.interest_rate

amount = 1000

for i, y in enumerate(range(number_of_years), start=1):
    amount += interest_rate/100*amount
    amount = round(amount, 2)
    print(f'Savings after {i:2d} years: {amount:.2f}')
```
