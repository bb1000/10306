```python
%%file test_leap_parametrized.py
import pytest
from leap import is_leap_year

@pytest.mark.parametrize('year, expected', [(2000, True), (1999, False), (1998, False), (1996, True), (1900, False), (1800, False), (1600, True)])
def test_leap_year(year, expected):
    assert is_leap_year(year) == expected
    #assert is_leap_year(2000) == True
    #assert is_leap_year(1999) == False
    #ssert is_leap_year(1998) == False
    #assert is_leap_year(1996) == True
    #assert is_leap_year(1900) == False
    #assert is_leap_year(1800) == False
    #assert is_leap_year(1600) == True
```

* handle turn of centuries

```python
%%file leap.py
def is_leap_year(year):
    if year % 100 == 0:
        if year % 400 == 0:
            return True
        else:
            return False
    else:        
        if year % 4 == 0:
            return True
        else:
            return False
```

```python
!pytest test_leap_parametrized.py
```

* refactoring (improve code without changing functionality)

```python
%%file leap.py
def is_leap_year(year):
    if year % 100 == 0:
        return year % 400 == 0
    else: 
        return year % 4 == 0
```
