## Example: savings calculator

* calculate savings at retirement age
* given a starting age, stop age
* given an interest

```python
%%file savings.py
def calculate_savings(start_age: int, stop_age: int, annual: int, interest: float) -> list[int]:
    savings_history = []
    current_savings = 0
    for age in range(start_age, stop_age):
        current_savings += annual
        current_savings += round(current_savings*interest/100)
        savings_history.append(current_savings)

    return savings_history
    
    
    ...
```

```python
%%file test_savings.py
from savings import calculate_savings

def test_one_year_10_percent():
    result = calculate_savings(start_age=19, stop_age=20, annual=6000, interest=10)
    assert result[0] == 6600, f'expected=6600 actual={result[0]}'

def test_three_years_0_percent():
    result = calculate_savings(start_age=19, stop_age=22, annual=6000, interest=0)
    assert result == [6000, 12000, 18000]
```
